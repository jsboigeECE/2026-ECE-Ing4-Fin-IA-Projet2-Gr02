# Data Preprocessing and Imbalance Analysis for Credit Card Fraud Detection

Phase 1.2: Preprocessing and Phase 1.3: Imbalance Analysis

```python
# Import Required Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import os
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Dataset
file_path = 'dataset/creditcard.csv'
if not os.path.exists(file_path):
    raise FileNotFoundError(f"Dataset not found at {file_path}")

df = pd.read_csv(file_path)
print(f"✅ Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
print(df.head())
```

```python
# Normalize Amount and Time Columns
features_to_scale = ['Amount', 'Time']
scaler = StandardScaler()
X_scaled = df.copy()
X_scaled[features_to_scale] = scaler.fit_transform(df[features_to_scale])
print("✅ Features scaled:", features_to_scale)
```

```python
# Split Data into Train/Test Sets
X = X_scaled.drop('Class', axis=1)
y = X_scaled['Class']
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
print("✅ Preprocessing completed:")
print(f"   Train set: {X_train.shape[0]} samples")
print(f"   Test set: {X_test.shape[0]} samples")
```

```python
# Save Preprocessed Data
output_dir = 'results/'
os.makedirs(output_dir, exist_ok=True)

# Save data
np.savez_compressed(f'{output_dir}/preprocessed_data.npz',
                   X_train=X_train.values,
                   X_test=X_test.values,
                   y_train=y_train.values,
                   y_test=y_test.values)

# Save scaler
joblib.dump(scaler, f'{output_dir}/scaler.pkl')

print(f"✅ Preprocessed data saved to {output_dir}")
```

```python
# Perform Imbalance Analysis
fraud_count = y.sum()
total_count = len(y)
fraud_ratio = fraud_count / total_count
print(f"✅ Imbalance Analysis:")
print(f"   Total transactions: {total_count}")
print(f"   Fraud cases: {fraud_count}")
print(f"   Fraud ratio: {fraud_ratio:.6f} ({fraud_count}/{total_count})")
```

```python
# Visualize Class Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=y, palette='viridis')
plt.title('Class Distribution (0: Normal, 1: Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.yscale('log')  # Use log scale due to imbalance
plt.show()

# Summary
print("Summary of class distribution:")
print(y.value_counts())
```
