# Chargement des Données Prétraitées

Charger les datasets prétraitées et le scaler pour la Phase 2

```powershell
import numpy as np
import joblib
import os

# Chemins
data_file = '../results/preprocessed_data.npz'
scaler_file = '../results/scaler.pkl'

# Charger les données
data = np.load(data_file)
X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']

# Charger le scaler
scaler = joblib.load(scaler_file)

print("✅ Données chargées avec succès!")
```

## Vérification des Dimensions

```powershell
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

print(f"\nFraudes en train: {y_train.sum()} ({y_train.sum()/len(y_train)*100:.3f}%)")
print(f"Fraudes en test: {y_test.sum()} ({y_test.sum()/len(y_test)*100:.3f}%)")
```

## Infos sur le Scaler

```powershell
print(f"Scaler type: {type(scaler)}")
print(f"Mean: {scaler.mean_}")
print(f"Scale: {scaler.scale_}")
```
